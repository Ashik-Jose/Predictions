```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('canada_per_capita_income.csv')
data
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

train = model.fit(data[['year']], data[['per capita income (US$)']])
```

```python
plt.scatter(data[['year']], data[['per capita income (US$)']],marker='+', color='red')

plt.plot(data[['year']], train.predict(data[['year']]), color='blue')

plt.show()
```

```python
model.predict([[2025]])
```

```python
model.score(data[['year']], data[['per capita income (US$)']])
```
